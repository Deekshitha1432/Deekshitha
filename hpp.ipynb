## Train_Test_split

```python
from sklearn.model_selection import train_test_split
```

```python
X = final_df_train.drop(['SalePrice'],axis=1)
y = final_df_train.drop(X,axis=1)
```

```python
final_df_train.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

## Model Building For Train Data

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
ridge = Ridge(alpha=1.0,random_state=42)  # alpha is the regularization strength
ridge.fit(X_train, y_train)
```

```python
y_pred = ridge.predict(X_train)
```

```python
mse = mean_squared_error(y_train,y_pred)
mae = mean_absolute_error(y_train,y_pred)
r2 = r2_score(y_train,y_pred)
rmse = np.sqrt(mean_squared_error(y_train, y_pred))
print("Mean Absolute error:", mae)
print("Mean squared error:", mse)
print("Root Mean Squared Error:", rmse)
print("R-squared:", r2)
```

```python
ridge.fit(X_test, y_test)
```

```python
y_preds =ridge.predict(X_test)
```

```python
mae = mean_absolute_error(y_test, y_preds)
mse = mean_squared_error(y_test, y_preds)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_preds)
print("Mean Absolute Error:",mae)
print("Mean Squared Error:", mse)
print("Root Mean Squared Error:", rmse)
print("R-squared:", r2)
```

## Grid search cv to find optimal parameter

```python
##GridSearchCV to find optimal hyperparameters
from sklearn.model_selection import GridSearchCV
param_grid = {'alpha': [0.001, 0.01, 0.1, 1, 10, 100]}
ridge = Ridge()
ridge_cv = GridSearchCV(ridge, param_grid, cv=5)
ridge_cv.fit(X_train, y_train)
print("Best alpha parameter: ", ridge_cv.best_params_)
```
